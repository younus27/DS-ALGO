```python
class Book:
    def __init__(self, pages, price, author, id, title):
        self.id     = id
        self.title  = title
        self.price  = price
        self.pages  = pages
        self.author = author
```

```python
import collections

class BookStore:
    def __init__(self, bookStoreName, Bookdictionary):
        self.bookStoreName  = bookStoreName
        self.Bookdictionary = Bookdictionary
        
    def searchBookBypages(self, n):
        matched = [ obj for obj in self.Bookdictionary if obj.pages == n]
#         print("matched:\t",matched)
        if len(matched) == 0:
            print("No Data Found")
            return
        for obj in  matched:
            print(obj.pages)
            print(obj.price)
            print(obj.author)
            print(obj.id)
            print(obj.title)
        
    def countBookByauthor(self):
        authors = [ obj.author for obj in self.Bookdictionary]
#         print("Authors:\t",authors)
        if len(authors) == 0:
            print("No Data Found")
            return
        dict_ = collections.Counter(authors)
        for key, value in dict_.items():
            print(key,value)
```

```python
def main():
    n = int(input())
    list_of_books = []
    
    for i in range(n):
        pages = int(input())
        price = int(input())
        author = input()
        id = int(input())
        title = input()
        
        obj = Book(pages, price, author, id, title)
        list_of_books.append(obj)

        
    bStore = BookStore("MyBookStore", list_of_books)
    
    no_of_pages = int(input())
    
    bStore.searchBookBypages(no_of_pages)
    
    bStore.countBookByauthor()

if __name__ == "__main__":
    main()
```
